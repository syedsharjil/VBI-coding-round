```python
num2words1 = {1: 'One', 2: 'Two', 3: 'Three', 4: 'Four', 5: 'Five', \
            6: 'Six', 7: 'Seven', 8: 'Eight', 9: 'Nine'}
def get_number_text(Number):
    if 0 <= Number <= 9:
        return num2words1[Number]
    else:
        print('Number out of implemented range of numbers.')
```

```python
word = "1145221"
count=1
length=""
if len(word)>1:
    for i in range(1,len(word)):
        if word[i-1]==word[i]:
            count+=1
        else:
            length += get_number_text(count)+ ' ' + word[i-1]+"'s "+" and "
            count=1
    length += (get_number_text(count) + ' ' + word[i]+ "'s")
else:
    i=0
    length += (get_number_text(count) + ' ' + word[i]+"'s" + " and ")
print (length)
```
